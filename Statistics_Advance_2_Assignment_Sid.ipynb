#Statistics Advance Theoretical Questions

####1. What is hypothesis testing in statistics ?
-  Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves making an assumption (hypothesis), collecting data, and determining whether the data provides enough evidence to reject that assumption.

####2. What is the null hypothesis, and how does it differ from the alternative hypothesis ?
-  The **null hypothesis (H₀)** is the default or original assumption, often stating there is no effect or no difference. The **alternative hypothesis (H₁ or Ha)** is what you aim to support, indicating a significant effect or difference. They are opposites—rejecting the null supports the alternative.

####3. What is the significance level in hypothesis testing, and why is it important ?
-  The significance level (α) is the threshold for deciding when to reject the null hypothesis, commonly set at 0.05 (5%). It represents the probability of making a Type I error—rejecting a true null hypothesis. It's important because it controls the rate of false positives.

####4. What does a P-value represent in hypothesis testing ?
-  A P-value is the probability of obtaining results as extreme as (or more extreme than) the observed data, assuming the null hypothesis is true. It tells us how compatible the data is with the null hypothesis.

####5. How do you interpret the P-value in hypothesis testing ?
-  If **P-value ≤ α**, reject the null hypothesis (significant result).  

If **P-value > α**, fail to reject the null hypothesis (not significant).  
Smaller P-values indicate stronger evidence against the null.

####6. What are Type 1 and Type 2 errors in hypothesis testing ?
-  **Type I Error (α):** Rejecting a true null hypothesis (false positive).  

**Type II Error (β):** Failing to reject a false null hypothesis (false negative).  

Balancing these is key in designing tests.

####7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?
-  **One-tailed test:** Tests for an effect in one direction (e.g., greater than).  

**Two-tailed test:** Tests for an effect in either direction (e.g., different from).  

####8. What is the Z-test, and when is it used in hypothesis testing ?
-  A Z-test is used to determine if there is a significant difference between sample and population means when the population standard deviation is known and the sample size is large (n > 30).

####9. How do you calculate the Z-score, and what does it represent in hypothesis testing ?
-  The Z-score is a statistical measure that tells you how many standard deviations a data point (or sample statistic) is from the mean. It plays a key role in hypothesis testing, especially when you're dealing with normally distributed data or large sample sizes.

The Z-score shows how many standard errors the sample mean is from the population mean.

####10. What is the T-distribution, and when should it be used instead of the normal distribution ?
-  The T-distribution is used when the sample size is small (n < 30) and the population standard deviation is unknown. It is similar to the normal distribution but has heavier tails.

####11. What is the difference between a Z-test and a T-test ?
-  **Z-test:** Population standard deviation known, large sample size.  

**T-test:** Population standard deviation unknown, small sample size.  
T-test is more conservative with smaller samples.

####12. What is the T-test, and how is it used in hypothesis testing ?
-  A T-test compares means to determine if they are significantly different.

Types include:
  - **One-sample T-test:** Sample vs. known value.  
  - **Two-sample T-test:** Two independent samples.  
  - **Paired T-test:** Same group tested twice.  

####13. What is the relationship between Z-test and T-test in hypothesis testing ?
-  Both test hypotheses about means. The **Z-test** is a special case of the T-test when the population standard deviation is known and/or the sample size is large. The **T-test** adapts to smaller samples with unknown variance.

####14. What is a confidence interval, and how is it used to interpret statistical results ?
-  A confidence interval (CI) gives a range of values likely to contain the population parameter. For example, a 95% CI means we are 95% confident the true value lies within that range. It helps assess estimate precision.

####15. What is the margin of error, and how does it affect the confidence interval ?
-  The margin of error (MoE) is the range added/subtracted from a statistic to form the CI. Larger MoE = wider CI = less precision. It's influenced by sample size and variability.

####16. How is Bayes' Theorem used in statistics, and what is its significance ?
-  Bayes' Theorem updates the probability of a hypothesis based on new evidence. It’s crucial in **Bayesian statistics**, where probabilities represent degrees of belief and are updated as more data becomes available.

####17. What is the Chi-square distribution, and when is it used ?
-  The Chi-square (χ²) distribution is used in categorical data analysis, particularly for variance estimation and testing relationships between variables (e.g., independence). It’s skewed and only takes non-negative values.

####18. What is the Chi-square goodness of fit test, and how is it applied ?
-  The Chi-square goodness of fit test assesses whether an observed frequency distribution matches an expected distribution. Used to test if a single categorical variable fits a theoretical model.

####19. What is the F-distribution, and when is it used in hypothesis testing ?
-  The F-distribution is used to compare two variances. It’s the basis for **ANOVA** and the **F-test**. It's right-skewed and used when testing multiple groups.

####20. What is an ANOVA test, and what are its assumptions ?
-  ANOVA (Analysis of Variance) compares means across three or more groups.

Assumptions:
  - Independent samples  
  - Normally distributed populations  
  - Equal variances (homogeneity of variance)

####21. What are the different types of ANOVA tests ?
-  Types of ANOVA:
  - **One-way ANOVA:** One independent variable.  
  - **Two-way ANOVA:** Two independent variables.  
  - **Repeated measures ANOVA:** Same subjects across multiple conditions.  
  - **MANOVA:** Multiple dependent variables.

####22. What is the F-test, and how does it relate to hypothesis testing ?
-  The F-test compares variances to determine if they are significantly different. It's central to **ANOVA**, where it assesses whether group means differ more than expected by chance.

```python
#17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
import scipy.stats as stats

# Step 1: Generate random data from a normal distribution
np.random.seed(42)
population_mean = 50
population_std = 10
sample_size = 100
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: Calculate sample mean and standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)

# Step 3: Calculate the confidence interval for the mean (95% confidence level)
confidence_level = 0.95
alpha = 1 - confidence_level
degrees_of_freedom = sample_size - 1

# t critical value for 95% confidence interval
t_critical = stats.t.ppf(1 - alpha/2, degrees_of_freedom)

# Margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

# Confidence Interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 4: Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"95% Confidence Interval for the Mean: ({ci_lower:.2f}, {ci_upper:.2f})")

# Interpretation
print("\nInterpretation: The true population mean is expected to lie within this confidence interval 95% of the time.")
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

# Sample data (e.g., test scores from a class)
sample = np.array([72, 75, 78, 70, 74, 73, 77, 76, 71, 74])

# Known population parameters
population_mean = 70
population_std = 5  # Known population standard deviation

# Sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Z-test statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output the results
print(f"Sample Mean = {sample_mean}")
print(f"Z-Score = {z_score}")
print(f"P-Value = {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

def plot_errors(alpha=0.05, beta=0.2):
    # Type I and Type II error regions
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.plot(x, y, label='Standard Normal Distribution')
    plt.fill_between(x, 0, y, where=(x > 1.96), color='red', alpha=0.3, label="Type I Error (alpha)")
    plt.fill_between(x, 0, y, where=(x < -1.96), color='teal', alpha=0.3, label="Type II Error (beta)")

    plt.axvline(x=1.96, color='black', linestyle='--', label='Critical Value (alpha=0.05)')
    plt.title('Type 1 and Type 2 Errors')
    plt.legend()
    plt.show()

plot_errors()
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

def confidence_interval(data, confidence_level=0.95):
    mean = np.mean(data)
    std_dev = np.std(data)
    n = len(data)
    z_score = stats.norm.ppf((1 + confidence_level) / 2)

    margin_of_error = z_score * (std_dev / np.sqrt(n))
    return mean - margin_of_error, mean + margin_of_error

# Example
sample_data = np.random.normal(loc=100, scale=15, size=50)
ci = confidence_interval(sample_data)
print(f"Confidence Interval: {ci}")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

def margin_of_error(data, confidence_level=0.95):
    mean = np.mean(data)
    std_dev = np.std(data)
    n = len(data)
    z_score = stats.norm.ppf((1 + confidence_level) / 2)

    return z_score * (std_dev / np.sqrt(n))

# Example
sample_data = np.random.normal(loc=100, scale=15, size=50)
margin = margin_of_error(sample_data)
print(f"Margin of Error: {margin}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

# Bayes' Theorem: P(A|B) = (P(B|A) * P(A)) / P(B)
def bayes_theorem(prior, likelihood, evidence):
    return (likelihood * prior) / evidence

prior = 0.2  # Prior probability
likelihood = 0.9  # Likelihood of observing evidence
evidence = 0.5  # Total probability of evidence

posterior = bayes_theorem(prior, likelihood, evidence)
print(f"Posterior Probability: {posterior}")
```

```python
#16. Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

# Step 1: Simulate two datasets with different variances
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=100)  # Variance = 100
group2 = np.random.normal(loc=50, scale=5, size=100)   # Variance = 25

# Step 2: Calculate the variances of both groups
var_group1 = np.var(group1, ddof=1)
var_group2 = np.var(group2, ddof=1)

# Step 3: Perform the F-test for variances
f_stat = var_group1 / var_group2  # F-statistic: larger variance / smaller variance
dfn = len(group1) - 1  # Degrees of freedom for numerator (group1)
dfd = len(group2) - 1  # Degrees of freedom for denominator (group2)

# Calculate the p-value using the F-distribution
p_value = 1 - f.cdf(f_stat, dfn, dfd)

# Step 4: Output the results
print(f"Variance of Group 1: {var_group1:.2f}")
print(f"Variance of Group 2: {var_group2:.2f}")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")

# Step 6: Visualize the distributions and variances
plt.figure(figsize=(8, 5))

# Plot histogram for Group 1
plt.hist(group1, bins=15, alpha=0.7, label='Group 1 (Variance=100)', color='teal', edgecolor='black')

# Plot histogram for Group 2
plt.hist(group2, bins=15, alpha=0.7, label='Group 2 (Variance=25)', color='yellow', edgecolor='black')

plt.title('Comparison of Group Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
